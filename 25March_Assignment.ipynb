# Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

# Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.

- Yes, there are outliers in the dataset.

- Outliers are points that are located outside the whiskers of the plot, which are typically defined as 1.5 times the interquartile range (IQR) below the first quartile or above the third quartile.

- Outliers can skew statistical measures like the mean and standard deviation, making them less representative of the central tendency and variability of the majority of the data.

- Outliers impact our analysis by skewing the result and affecting the accuracy.

# Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different app categories.

```python
import matplotlib.pyplot as plt

df1 = df1[df1['Rating'] <= 5]
df1.boxplot(column='Rating', by='Category', figsize=(20, 15), color='blue',rot=90)
plt.xlabel('Category')
plt.ylabel('Rating')
plt.title('Comparison of Rating by Category')

plt.show()
```

# Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 18))
plt.scatter(df1['Size'], df1['Rating'], c=df1['Rating'],cmap='viridis')
plt.xticks(rotation=90)
plt.xlabel('Size')
plt.ylabel('Rating')
plt.title('Relationship between App Size and Rating')

plt.show()
```

# Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Change the Price data to float value
df1['Price'] = df1['Price'].str.split('$').str[1].astype(float)

# Group the dataset by app type and calculate the average price
grouped = df1.groupby('Type')['Price'].mean()
grouped.fillna(0, inplace=True)

# Create a bar chart
plt.bar(grouped.index, grouped.values)
plt.xlabel('Type')
plt.ylabel('Average Price')
plt.title('Average Prices by App Type')

# Show the plot
plt.show()
```

# Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
df1['Installs'] = df1['Installs'].str.replace(',', '')
df1['Installs'] = df1['Installs'].str.replace('+', '').astype(int)

freq_table = df1.sort_values(by=['Installs'], ascending=False)
freq_table[['App', 'Installs']].reset_index(drop=True).head(10)
```

# Q15. A company wants to launch a new app on the Google Playstore and has asked you to analyze the Google Playstore dataset to identify the most popular app categories. How would you approach this task, and what features would you analyze to make recommendations to the company?

```python
frequency_table = pd.DataFrame(df1.groupby(['Category'])['Installs'].sum().sort_values(ascending=False).reset_index())
frequency_table
```

```python
frequency_table.boxplot(column='Installs', by='Category', figsize=(20, 12), color='blue', rot=90)
plt.xlabel('Category')
plt.ylabel('Installs')
plt.title('Comparison of total installs by Category')

plt.show()
```

- By using this features we will find most popular app category.
